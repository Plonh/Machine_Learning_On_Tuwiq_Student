### Label encoding:

```python
For_label_col = ['Home Region','Home City','Program Main Category Code',
                     'Program Sub Category Code','College']
train_df[For_label_col]
```

```python
for col in For_label_col:
    train_df[col].fillna(method='bfill', inplace=True)
    train_df[col].fillna(method='ffill', inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in For_label_col:
    label_encoders[col] = LabelEncoder()
    train_df[col] = label_encoders[col].fit_transform(train_df[col])

train_df[For_label_col]
```

### Combine the encoded columns with the training dataset:

```python
y = train_df['Y']
post_df = pd.concat([train_df[num_cols].copy(),train_df[For_label_col],oneHot_df,y],axis=1)
X = post_df.loc[:, post_df.columns != 'Y']
```

### Model building:

```python
# Import Libraries
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Training by LogisticRegression
lg = LogisticRegression()
lg.fit(X,y)
```

```python
# Predict model
y_pred = lg.predict(X)
```

```python
# Evaluate the performance
accuracy = accuracy_score(y,y_pred)
precision = precision_score(y,y_pred)
recall = recall_score(y,y_pred)
f1 = f1_score(y,y_pred)
```

```python
print("accuracy: ", accuracy)
print("precision: ", precision)
print("recall: ", recall)
print("f1: ", f1)
```

```python
# Import Libraries
from sklearn.tree import DecisionTreeClassifier
```

```python
# Training by DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
tree_model.fit(X, y)
```

```python
# Predict model
tree_y_pred = tree_model.predict(X)
```

```python
# Evaluate the performance
tree_accuracy = accuracy_score(y,tree_y_pred)
tree_precision = precision_score(y,tree_y_pred)
tree_recall = recall_score(y, tree_y_pred)
tree_f1 = f1_score(y, tree_y_pred)
```

```python
print("tree_accuracy: ", tree_accuracy)
print("tree_precision: ", tree_precision)
print("tree_recall: ", tree_recall)
print("tree_f1: ", tree_f1)
```

```python
# Import Libraries
from sklearn.ensemble import RandomForestClassifier
```

```python
# Training by RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X, y)
```

```python
# Predict classes model
rf_y_pred = rf_model.predict(X)
```

```python
# Evaluate the performance
rf_accuracy = accuracy_score(y, rf_y_pred)
rf_precision = precision_score(y, rf_y_pred)
rf_recall = recall_score(y, rf_y_pred)
rf_f1 = f1_score(y, rf_y_pred)
```

```python
print("rf_accuracy: ", rf_accuracy)
print("rf_precision: ", rf_precision)
print("rf_recall: ", rf_recall)
print("rf_f1: ", rf_f1)
```
